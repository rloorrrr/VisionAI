```python
#허깅페이스에서 데이터셋 로드하기
from datasets import load_dataset

ds = load_dataset("yesidobyte/nsfw1024")
```

```python
from PIL import Image, ImageFile
import os

# 손상된 이미지를 건너뛰도록 설정
ImageFile.LOAD_TRUNCATED_IMAGES = True

# 이미지 파일을 저장할 디렉토리 설정
output_dir = "/content/drive/MyDrive/vs/images"
os.makedirs(output_dir, exist_ok=True)

# 이미지 파일을 저장할 때 이름을 604부터 시작하도록 설정
start_index = 604

for i, example in enumerate(ds['train'], start=start_index):  # start를 604로 설정
    try:
        image = example['image']
        # 파일명에 604부터 시작하도록 설정
        image_filename = f"nsfw_{i}.png"
        image.save(os.path.join(output_dir, image_filename))
    except Exception as e:
        print(f"Error with image {i}: {e}")
        continue

print("Images have been saved, skipping corrupted ones.")
```
