```python
from datasets import load_dataset
from PIL import Image, ImageFile
import os

# 손상된 이미지를 건너뛰도록 설정
ImageFile.LOAD_TRUNCATED_IMAGES = True

# 데이터셋 스트리밍으로 로드
streaming_ds = load_dataset("amaye15/NSFW", split="train", streaming=True)

# 이미지 파일을 저장할 디렉토리 설정
output_dir = "/home/work/hi/data/images"
os.makedirs(output_dir, exist_ok=True)

# 이미지를 저장할 때 이름을 4893부터 시작하도록 설정
start_index = 4893

# 스트리밍 데이터셋에서 이미지 저장
for i, example in enumerate(streaming_ds, start=start_index):
    if i >= start_index + 15000:  # 15,000개 처리 후 종료
        break
    try:
        # pixel_values를 PIL 이미지로 바로 사용
        image = example["pixel_values"]

        # 파일명 생성
        image_filename = f"nsfw_{i}.png"
        image.save(os.path.join(output_dir, image_filename))

    except Exception as e:
        print(f"Error with image {i}: {e}")
        continue

print("Images have been saved, skipping corrupted ones.")
```
